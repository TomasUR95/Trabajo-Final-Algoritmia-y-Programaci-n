```python
import pandas as pd
import random

# Carga los apellidos desde el archivo CSV
apellidos = pd.read_csv('https://raw.githubusercontent.com/TomasUR95/Trabajo-Final-Algoritmia-y-Programaci-n/main/Datasets/apellidos.csv')
apellidos
```

```python
# Verifica los apellidos de las columnas
print(apellidos.columns)

# Convierte la primera columna a una lista
apellidos_lista = apellidos.iloc[:, 0].tolist()
```

```python
# Carga los nombres desde el archivo CSV
nombres = pd.read_csv('https://raw.githubusercontent.com/TomasUR95/Trabajo-Final-Algoritmia-y-Programaci-n/main/Datasets/nombres.csv')
nombres
```

```python
 # Verifica los nombres de las columnas
print(nombres.columns)

# Convierte la primera columna a una lista
nombres_lista = nombres.iloc[:, 0].tolist()
```

```python
# Lista para almacenar los nombres completos de los estudiantes
estudiantes = set()

# Mientras no tengamos 1000 estudiantes únicos
while len(estudiantes) < 1000:
    # Selecciona aleatoriamente dos nombres y dos apellidos
    nombre1 = random.choice(nombres_lista)
    nombre2 = random.choice(nombres_lista)
    apellido1 = random.choice(apellidos_lista)
    apellido2 = random.choice(apellidos_lista)

    # Crea el nombre completo
    nombre_completo = f"{nombre1} {nombre2} {apellido1} {apellido2}"

    # Agrega el nombre completo al conjunto si no está duplicado
    estudiantes.add(nombre_completo)

# Convierte el conjunto a una lista
estudiantes = list(estudiantes)

# Verifica que se hayan generado 1000 nombres únicos
print(f"Número de estudiantes generados: {len(estudiantes)}")

# Imprime los nombres de los estudiantes
for estudiante in estudiantes:
    print(estudiante)
```
